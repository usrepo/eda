**task-1**
- Create two lists
- One is categorical column list
- Another is numercial column list
- using dtypes

**HINT**

- Whenever you want to create any list from series
- First convert the series into dictionary
- Iterate through dictionary
- If that value == object ytpe, then keep that corresponding key in one list
- also keep the variable in another list

```python
data_type = visa_df.dtypes
dict(data_type)
```

```python
cat_list = []
num_list = []
for x,y in dict(data_type).items():
    if y == 'object':
        cat_list.append(x)
    else:
        num_list.append(x)
print(cat_list)
print(num_list)
```

```python
cat_list = [x for x,y in dict(data_type).items() if y=='object']
num_list = [x for x,y in dict(data_type).items() if y!='object']
cat_list, num_list
```

```python
len(cat_list), len(num_list)
```

- We cannot create a dataframe using these two lists
- because these tow lists are not in equal length
- If you want to create a dataframe, 
- then we need to make two lists has equal length
    - then you need to question yourself
    - making equal length required or not
    - if it is require how?

**Create columns dataframe**

```python
num_df = pd.DataFrame(num_list, columns=['Numerical_Column'])
num_df.to_csv('NumCol_df.csv',index=False)
```

```python
num_df = pd.DataFrame(cat_list, columns=['Categorical_Column'])
num_df.to_csv('CatCol_df.csv',index=False)
```

```python
cols,data_types=[],[]
for keys, values in dict(visa_df.dtypes).items():
    cols.append(keys)
    data_types.append(values)
    
pd.DataFrame(zip(cols,data_types), columns=['Columns', 'Data Type'])
```

$take$

- first argument: indices
    - indices will take some values in array which means in a list
    - it tries to extract the data
    - that values we provided represents columns or rows?
- second argument: axis
    - 0 represents row (default)
    - 1 represents column


```python
#Case-1
visa_df.take([100,200,300], axis=0)
# it retrieves 3 rows which having index 100, 200, 300
```

```python
visa_df.take([100,200,300], axis=1)
# this will check for 100th, 200th andf 300th column
# but originally we have 12 columns (0 to 11)
# so we will get error IndexError: indices are out-of-bounds
```

```python
#Case-2
visa_df.take([1,4,6], axis=1)
# the output also behave as dataframe
```

```python
#Case-3
visa_df.take([1,4,6], axis=1).head()
```
